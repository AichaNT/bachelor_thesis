```python
import pandas as pd
import json
```

```python

# --- Step 1: Load original CSVs ---
springer_orig1 = pd.read_csv("data/raw/manual_search/Springer/springer_query_-2024_2026-03-02.csv")
springer_orig2 = pd.read_csv("data/raw/manual_search/Springer/springer_query_2025-2026_2026-03-02.csv")

springer_original = pd.concat([springer_orig1, springer_orig2], ignore_index=True)

# Extract and clean DOIs
orig_dois = (
    springer_original["Item DOI"]
    .dropna()
    .astype(str)
    .str.lower()
    .str.strip()
)
orig_dois = set(orig_dois)
print(f"Total original DOIs: {len(orig_dois)}")

# --- Step 2: Load JSON metadata ---
json_path = "data/raw/API_search/springer_api_metadata.JSON"
json_dois = set()

with open(json_path) as f:
    for line in f:
        obj = json.loads(line)
        for rec in obj.get("records", []):
            doi = rec.get("doi", "").lower().strip()
            if doi:
                json_dois.add(doi)

print(f"Total DOIs in JSON: {len(json_dois)}")

# --- Step 3: Identify missing DOIs ---
missing_dois = orig_dois - json_dois
print(f"DOIs missing in JSON: {len(missing_dois)}")

# Optional: save missing DOIs to a CSV for later fetching
missing_df = pd.DataFrame({"doi": list(missing_dois)})
missing_df.to_csv("data/processed/missing_springer_dois.csv", index=False)
print("Missing DOIs saved to data/processed/missing_springer_dois.csv")
```
